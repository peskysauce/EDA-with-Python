## Reformatting column names

```python
df = df.rename(columns={'Formatted Date': 'date', 
                   'Summary': 'summary',
                   'Precip Type': 'precip',
                   'Temperature (C)': 'temp_c',
                   'Apparent Temperature (C)': 'app_temp_c',
                   'Humidity': 'humidity',
                   'Wind Speed (km/h)': 'wind_speed_kmh',
                   'Wind Bearing (degrees)': 'wind_bearing',
                   'Visibility (km)': 'visibility_km',
                   'Loud Cover': 'loud_cover',
                   'Pressure (millibars)': 'pressure_mbar',
                   'Daily Summary': 'daily_summary',
})

df.head(100)
```

```python
# Loud_cover doesn't convey any meaningful information and only contains value of 0 for the entire dataset so I will drop the column
# As for summary I will use the 'summary' column and drop 'daily_summary'
df = df.drop(['loud_cover', 'daily_summary'], axis=1)

df.head()
```

```python
# I want to split datetime into day, month and year columns. But I have to first parse the column as it is currently being stored as an object type
df.date.dtypes

df['date'].str.len().unique() # Date is consistent

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S.%f', utc=True)
```

```python
df.date.head()
```

```python
df['day'] = pd.to_datetime(df['date']).dt.day
df['month'] = pd.to_datetime(df['date']).dt.month
df['year'] = pd.to_datetime(df['date']).dt.year
df = df.drop(columns='date')
df
```

## Cleaning up outliers

We know that the distribution for these variables above is skewed, so to remove outliers we will use the IQR method instead of z-score.

For pressure_mbar column we will drop anything below 100 because there seemed to be some 0 values

```python
iqr = df.quantile(0.75) - df.quantile(0.25)
up_bound = df.quantile(0.75) + (iqr * 1.5)
low_bound = df.quantile(0.25) - (iqr * 1.5)

up_bound
low_bound
```

```python
# Counting the total rows to drop

df[(df['humidity'] <= low_bound.humidity) | (df['humidity'] >= up_bound.humidity) | 
   (df['wind_speed_kmh'] <= low_bound.wind_speed_kmh) | (df['wind_speed_kmh'] >= up_bound.wind_speed_kmh) | 
   (df['pressure_mbar'] <= 100)
].count()

df.shape
# The total columns to drop will be 4261, we currently have 96416 rows
```

```python
df = df.drop(df[(df['humidity'] <= low_bound.humidity) | (df['humidity'] >= up_bound.humidity) | 
                (df['wind_speed_kmh'] <= low_bound.wind_speed_kmh) | (df['wind_speed_kmh'] >= up_bound.wind_speed_kmh) | 
                (df['pressure_mbar'] <= 100)
               ].index)

df.shape
```

```python
fig, [[ax1, ax2, ax3],[ax4, ax5, ax6]] = plt.subplots(2, 3, figsize=(18,10))
plt.tight_layout

sns.histplot(data=df, x='humidity', ax=ax1)
sns.histplot(data=df, x='wind_speed_kmh',ax=ax2)
sns.histplot(data=df, x='pressure_mbar',ax=ax3)
sns.boxplot(data=df, x='humidity', ax=ax4)
sns.boxplot(data=df, x='wind_speed_kmh', ax=ax5)
sns.boxplot(data=df, x='pressure_mbar', ax=ax6)
```

Graphs are looking much better, but there are still too many outliers on the pressure graph. Let's check the descriptive statistics for our pressure graph.

```python
df.pressure_mbar.describe()
```

Given that the graph is now closer to normally distributed, I will use z-score to eliminate the outliers, using 3 std.dev

```python
df.shape

df.pressure_mbar[(df['pressure_mbar'] <= (df.pressure_mbar.mean() - df.pressure_mbar.std()*3)) | (df['pressure_mbar'] >= (df.pressure_mbar.mean() + df.pressure_mbar.std()*3))].count()

df = df.drop(df.pressure_mbar[(df['pressure_mbar'] <= (df.pressure_mbar.mean() - df.pressure_mbar.std()*3)) | (df['pressure_mbar'] >= (df.pressure_mbar.mean() + df.pressure_mbar.std()*3))].index)

df.shape

sns.histplot(data=df, x='pressure_mbar')
```

## Regression and Questions to Answer

- Does temperature and humidity have any correlation?
- Does lower pressure always result in precipitation?
- Does lower pressure result in higher wind speed?
- Does higher humidity causes reduced visibility?
- What is the temperature trend throughout the year? Is 2016 hotter than 10 years ago?

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

### Temperature v Humidity

```python
X = df['temp_c']
Y = df['humidity']
X = sm.add_constant(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8)
model = sm.OLS(Y_train, X_train)    
result = model.fit()
print(result.summary2())
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.regplot(x=X_train['temp_c'], y=Y_train, marker='o', scatter_kws={'s':2, 'alpha': 0.3}, line_kws={"color": "orange"})
```

```python
Y_predict = result.predict(X_test)
test_result = (pd.DataFrame({'Actual': Y_test, 'Predicted': Y_predict})
               .reset_index()
               .drop(columns='index'))

RMSE = np.sqrt(mean_squared_error(Y_test, Y_predict))
r2 = r2_score(Y_test, Y_predict)

print('RMSE =', RMSE)
print('r2 =', r2)
```

```python
sns.lmplot(x='Actual', y='Predicted', data=test_result, legend=False, scatter_kws={'s':2, 'alpha': 0.8}, line_kws={"color": "orange"})
```

```python
# Residual plot
fig, ax = plt.subplots(figsize=(8,5))
sns.histplot(data=result.resid, alpha=0.4, color='teal')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.regplot(x=result.fittedvalues, y=result.resid, ax=ax, scatter_kws={'s':2, 'alpha': 0.8}, line_kws={"color": "orange"})
ax.set_xlabel('Predicted')
ax.set_ylabel('Residuals')
```

Temperature and humidity has weak to moderate negative relationship. In this data humidity takes continuous value between 0 to 1, which is 0% to 100%. So a unit increase in temperature lowers the humidity by approximately 1.81%, but subjected to bias based on how spread out the data points are as reflected by R-squared.

From our regression we can see that to a certain extent humidity is affected by temperature, 36.7% of the change in humidity to be exact. Our plot above also gave us a clue that while the data is relatively spread out, there is an area of the plot where data are most concentrated, between 20 degrees to 40 degrees celsius. The general trend here is as temperature continue to rise, humidity tends to fall.

By feeding our test data into our trained model, we could see a relatively wide error, which was expected given the relatively weak R-squared value. Root of MSE measures low difference so to a certain extend our model is fair for humidity prediction.
